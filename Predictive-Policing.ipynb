```python
import pandas as pd
import numpy as np
```

```python
# use the inline backend to generate the plots within the browser
%matplotlib inline 

import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.style.use('ggplot') # optional: for ggplot-like style

# check for latest version of Matplotlib
print ('Matplotlib version: ', mpl.__version__) # >= 2.0.0
```

```python
csv_path='D:\Documents\ISEN\Cours\ISEN\Informatique\DataScience\Projet\crime.csv'
df = pd.read_csv(csv_path, encoding='ISO-8859-1', dtype='unicode')
```

```python
df_hour = df.groupby('HOUR', axis=0).size()
df_hour = pd.DataFrame(df_hour)
df_hour.index = df_hour.index.astype(int, copy=True)
df_hour = df_hour.sort_index(ascending=True)
```

```python
df_hour.plot(kind='bar', figsize=(20, 10))

plt.title('Number of crime according to hour') # add a title to the histogram
plt.ylabel('Number of crimes') # add y-label
plt.xlabel('Hour') # add x-label

plt.show()
```

```python
df_day = df.groupby('DAY_OF_WEEK', axis=0).size()
df_day = pd.DataFrame(df_day)
df_day = df_day.reindex(["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"])
```

```python
df_day.plot(kind='bar', figsize=(20, 10))

plt.title('Number of crime according to day') # add a title to the histogram
plt.ylabel('Number of crimes') # add y-label
plt.xlabel('Day') # add x-label

plt.show()
```

```python
df_month = df.groupby('MONTH', axis=0).size()
df_month = pd.DataFrame(df_month)
df_month.index = df_month.index.astype(int, copy=True)
df_month = df_month.sort_index(ascending=True)
```

```python
df_month.plot(kind='bar', figsize=(20, 10))

plt.title('Number of crime according to month') # add a title to the histogram
plt.ylabel('Number of crimes') # add y-label
plt.xlabel('Month') # add x-label

plt.show()
```

```python
df_year = df.groupby('YEAR', axis=0).size()
df_year = pd.DataFrame(df_year)
df_year.index = df_year.index.astype(int, copy=True)
df_year = df_year.sort_index(ascending=True)
```

```python
df_year.plot(kind='bar', figsize=(20, 10))

plt.title('Number of crime according to year') # add a title to the histogram
plt.ylabel('Number of crimes') # add y-label
plt.xlabel('Year') # add x-label

plt.show()
```
